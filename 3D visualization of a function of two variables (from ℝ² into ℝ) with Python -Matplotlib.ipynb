Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math
from math import *
%matplotlib notebook
```

Let be a function f from ℝ² into ℝ. For any couple (x, y) in ℝ² :

```python
def f(x, y):
    return 2*x**2 - x*y + 2*y**2
```

We choose a range of values from the set of real numbers ℝ for x and y (I will take for the example [-100, 100] with 100 values for each variable).

```python
x = np.linspace(-100, 100, 100)
y = np.linspace(-100, 100, 100)
```

We will then create an array of values containing all value pairs (xi, yi) for xi in x, and yi in y.

```python
X, Y = np.meshgrid(x, y)
```

Then, we compute the values of our function defined above for each pair of values (xi, yi) and store these values in a new variable Z :


```python
Z = f(X,Y)
```

# Visualizations

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='jet', edgecolor = 'none')
```

You can go further and add a title, caption the axes, choose the size of the figure, play on the size of the cells of the figure (the rstride and cstride parameters), modify the cmap colour scale (I put a list of possible values for cmap at the bottom of the page in the bonus section), …

```python
# We define the size of the figure
fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='plasma_r', edgecolor='none')
# We caption the figure and set a size with fontsize
ax.set_title("Surface", fontsize = 13)
ax.set_xlabel('x', fontsize = 11)
ax.set_ylabel('y', fontsize = 11)
ax.set_zlabel('Z', fontsize = 10)
```

# Full code

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
def fonction1 (x, y):
    return 2*x**2 - x*y**2 + 2*y**2
x = np.linspace(-400, 400, 100)
y = np.linspace(-400, 400, 100)
X, Y = np.meshgrid(x, y)
Z = fonction1(X,Y)
fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=5, cstride=5,
                cmap='cool')
ax.set_title("Surface Bonus", fontsize = 13)
ax.set_xlabel('x', fontsize = 11)
ax.set_ylabel('y', fontsize = 11)
ax.set_zlabel('Z', fontsize = 11)
plt.show()
```
